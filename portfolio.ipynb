## Prompt Engineering

### Stap 1

Het doel van deze prompt is om een stappenplan te verkrijgen voor het webscrapen van de HBO Kennisbank, zodat er uiteindelijk een gestructureerde dataframe kan worden opgebouwd die gebruikt kan worden voor het genereren van samenvattingen.

De prompt maakt gebruik van Goal-Oriented Prompting en Contextual Prompting door het einddoel (samenvattingen genereren) en de gewenste output (een dataframe) expliciet te benoemen. Daarnaast zorgt Chain-of-Thought Prompting ervoor dat het model een gestructureerd en volledig stappenplan genereert.

### Antwoord

De output van deze prompt is gebruikt als conceptuele leidraad voor het project. De beschreven stappen hebben geholpen om inzicht te krijgen in de opbouw van een scraping pipeline, maar zijn niet direct in code geïmplementeerd.

### Stap 2

Het doel van deze prompt is om de eerder verkregen stappen om te zetten naar concrete, werkende code, zodat de scraping pipeline daadwerkelijk uitgevoerd kan worden en data automatisch verzameld wordt.

De prompt gebruikt Role Prompting door het model in de rol van data scientist te plaatsen, wat leidt tot meer robuuste en praktijkgerichte code. Daarnaast zorgt Instruction-based Prompting ervoor dat de output direct wordt vertaald naar een concrete implementatie.

### Antwoord

De gegenereerde code is direct overgenomen als basisimplementatie van de scraper. Deze code is vervolgens uitgevoerd en gebruikt om een eerste dataset te genereren, zonder grote structurele aanpassingen.

### Prompt 3

Het doel van deze prompt is om de scraper te verbeteren door rekening te houden met de hiërarchische structuur van de website, zodat niet alleen lijstpagina’s maar ook de onderliggende detailpagina’s correct worden gescrapet.

De prompt gebruikt Iterative Prompting door voort te bouwen op eerdere output en deze te verfijnen op basis van nieuwe inzichten. Daarnaast wordt Contextual Prompting toegepast door de specifieke situatie (listing vs. detailpagina’s) te beschrijven, waardoor het model gerichter kan verbeteren.

### Antwoord

De aangepaste code uit deze prompt is gekopieerd en gebruikt ter vervanging van de eerdere scraper, zodat ook detailpagina’s meegenomen worden. Dit resulteerde in een completere dataset met informatie uit meerdere niveaus van de website.
